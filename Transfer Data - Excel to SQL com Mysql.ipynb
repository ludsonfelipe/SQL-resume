# Conectando Para Criar Querys

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="123456",
  database="bancoteste"
)

mycursor = mydb.cursor()
```

## Lendo tabelas do Excel e criando Tabelas Mysql

```python
tabela_clientes = pd.read_excel('C:/Users/felipe/Desktop/Analise De Dados/Dados/Tabela Clientes.xlsx')
```

```python
tabela_clientes.head(1)
```

```python
mycursor.execute(f"CREATE TABLE Clientes (\
                 codigo INT  PRIMARY KEY NOT NULL,\
                 primeiro_nome VARCHAR(255) NOT NULL,\
                 sobrenome VARCHAR(255) NOT NULL,\
                 email VARCHAR(255) NOT NULL,\
                 genero VARCHAR(1) NOT NULL,\
                 num_filhos INT NOT NULL,\
                 data_de_nascimento DATE\
                 )")
```

```python
tabela_produtos = pd.read_excel('C:/Users/felipe/Desktop/Analise De Dados/Dados/Tabela Produtos.xlsx')
```

```python
tabela_produtos.head(1)
```

```python
mycursor.execute(f"CREATE TABLE Produtos (\
                 sku INT  PRIMARY KEY NOT NULL,\
                 tipo VARCHAR(255) NOT NULL,\
                 marca VARCHAR(255) NOT NULL,\
                 custo_de_compra FLOAT NOT NULL,\
                 valor_de_venda FLOAT NOT NULL,\
                 cor VARCHAR(20)\
                 )")
```

```python
tabela_vendas = pd.read_excel('C:/Users/felipe/Desktop/Analise De Dados/Dados/Tabela Vendas.xlsx')
```

```python
tabela_vendas.sample(1)
```

```python
mycursor.execute(f"CREATE TABLE Vendas (\
                 sku INT NOT NULL,\
                 dia DATE NOT NULL,\
                 loja VARCHAR(255) NOT NULL,\
                 cliente INT NOT NULL,\
                 quantidade INT NOT NULL,\
                 forma_de_pagamento VARCHAR(20)\
                 )")
```

```python
mycursor.execute('SHOW TABLES')
```

```python
for x in mycursor:
    print(x)
```

## Inserindo os dados do Excel na tabela Mysql

```python
# Criando tuplas de clientes
lista_clientes_values = []
for i in tabela_clientes.iterrows():
    lista_clientes_values.append((i[1][0],i[1][1],i[1][2],i[1][3],i[1][4],i[1][5],i[1][6]))
```

```python
sql = "INSERT INTO clientes (codigo, primeiro_nome, sobrenome, email, genero, num_filhos, data_de_nascimento) \
VALUES (%s, %s, %s, %s, %s, %s, %s)"

mycursor.executemany(sql,lista_clientes_values) # EXECUTA VARIAS TUPLAS

mydb.commit()
```

```python
lista_produtos_values = []
for i in tabela_produtos.iterrows():
    lista_produtos_values.append((i[1][0],i[1][1],i[1][2],i[1][3],i[1][4],i[1][5]))
```

```python
sql = "INSERT INTO Produtos (sku, tipo, marca, custo_de_compra, valor_de_venda, cor) \
VALUES (%s, %s, %s, %s, %s, %s)"

mycursor.executemany(sql,lista_produtos_values) # EXECUTA VARIAS TUPLAS

mydb.commit()
```

```python
lista_vendas_values = []
for i in tabela_vendas.iterrows():
    lista_vendas_values.append((i[1][0],i[1][1],i[1][2],i[1][3],i[1][4],i[1][5]))
```

```python
sql = "INSERT INTO Vendas (sku, dia, loja, cliente, quantidade, forma_de_pagamento) \
VALUES (%s, %s, %s, %s, %s, %s)"

mycursor.executemany(sql,lista_vendas_values) # EXECUTA VARIAS TUPLAS

mydb.commit()
```
